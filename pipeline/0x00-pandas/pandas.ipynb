### ðŸ‘‰ 0. From Numpy


```python
def from_numpy(array):
    """
    ****************************************
    creates a pd.DataFrame from a np.ndarray
    ****************************************
    @array: np.ndarray for creation of the pd.DataFrame
    *** The columns of the pd.DataFrame should be labeled
        in alphabetical order and capitalized. There will
        not be more than 26 columns.
    Returns:
            the newly created pd.DataFrame
    """
    a=65
    alphab = np.array(list(map(chr, range(a, a+array.shape[1]))))
    df = pd.DataFrame(array)
    df.columns=alphab
    return df
```

### ðŸ‘‰ 2. From File


```python
def from_file(filename, delimiter):
    """
    ****************************************
    loads data from a file as a pd.DataFrame
    ****************************************
    @filename: is the file to load from
    @delimiter: is the column separator
    Returns:
            the loaded pd.DataFrame
    """
    with open(filename, 'r') as f:
        headers = list(f.readline().strip().split(delimiter))
        body = []
        for line in f.readlines():
            body.append(np.array(line.strip().split(delimiter)))
    df = pd.DataFrame(np.array(body))
    df.columns = headers
    return df
```
